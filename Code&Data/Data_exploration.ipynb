```python
# Import data and obtain descriptive statistics
import pandas as pd
df = pd.read_csv('Dataset.csv')
df.describe()
```

```python
# Box plot for rental demand distribution across the year
import plotly.plotly as py
import plotly.graph_objs as go
import plotly
import calendar
data = []
for i in range(12):
    data.append(go.Box(y=df[pd.to_datetime(df['Date']).dt.month == i+1]['rental_count'], 
                       name=calendar.month_name[i+1], 
                       showlegend=False ) )
layout = go.Layout(
    title='Rental demand distribution',
    xaxis=dict(
        title='Month',
        titlefont=dict(
            size=18,
        )
    ),
    yaxis=dict(
        title='Daily rental count',
        titlefont=dict(
            size=18,
        )
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='rental_demand_distribution')
```
